```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
# Ignore divide by 0 error
np.seterr(all='ignore')

import sys
sys.path.append('..')
from nptk.mappers import NeuronMapper

NeuronMapper.setPrecision(127)
```

```python
num_points = 1000

x = np.linspace(-1, 1, num_points)
actual = []
for gain in x:
    pn = NeuronMapper.build([gain])
    actual.append(pn.step([1]))

plt.figure()
plt.plot(x, actual)
plt.xlabel("Expected Gain")
plt.ylabel("Actual Gain")
```
