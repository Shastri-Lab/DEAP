```python
import sys
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
sys.path.append('..')
from deap.photonics import MRRTransferFunction, MRMTransferFunction
```

```python
phi = np.linspace(np.deg2rad(-180), np.deg2rad(180), 500)

plt.figure()
mrr = MRRTransferFunction(a=1, r1=0.95, r2=0.95)
plt.plot(phi, mrr.throughput(phi), label="$T_p$")
plt.plot(phi, mrr.dropput(phi), label="$T_d$")
plt.xlabel("$\phi [rad]$")
plt.ylabel("$I/I_o$")
plt.legend()
```

```python
Tp = np.linspace(0, 1, 1000)
phi = mrr.phaseFromDropput(Tp)
plt.figure()
plt.plot(phi, Tp)
plt.xlabel("$phi$")
plt.ylabel("$T_p$")

mrr.phaseFromDropput(0)
```

```python
phi = np.linspace(-np.deg2rad(90), np.deg2rad(90), 300)

plt.figure()
mrm = MRMTransferFunction(a=0.9, r=0.9)
plt.plot(phi, mrm.throughput(phi), label="$T_p$")
plt.xlabel("$\phi$")
plt.ylabel("$I/I_o$")
```

```python
Tp = np.linspace(0.001, 1, 1000)
phi = mrm.phaseFromThroughput(Tp)

plt.figure()
plt.plot(phi, Tp)
plt.xlabel("$phi$")
plt.ylabel("$T_p$")
```
