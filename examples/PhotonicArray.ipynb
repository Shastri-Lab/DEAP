```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt

import sys
sys.path.append('..')
from deap.mappers import PWBArrayMapper
```

```python
image = plt.imread("./img/bass.jpg")
greyscale = np.mean(image, axis=2)
plt.figure("Original Image, Greyscale")
plt.imshow(greyscale, cmap='gray')
```

```python
gaussian_kernel = \
    np.array([
        [1, 2, 1],
        [2, 4, 2],
        [1, 2, 1]]) * 1/16
    
pa1 = PWBArrayMapper.build(greyscale.shape, gaussian_kernel, 1)
convolved = pa1.step(greyscale)
plt.figure("Convolved Image with Gaussian Filter")
plt.imshow(convolved, cmap='gray')
```
